```python
# Transport + diffusion (non self-adjoint)
x, xi = symbols('x xi', real=True)
c = 5.0    # Convection speed
nu = 0.1   # Diffusion coefficient

symbol = -1j * c * xi - nu * xi**2  # -c∂_x - ν∂_xx

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-10, 10, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3, 1e-4],
    resolution=200,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Oscillateur amorti
x, xi = symbols('x xi', real=True)
omega = 3.0  # Frequency
gamma = 0.5  # Damping

symbol = xi**2 + 1j*gamma*xi + omega**2  # ∂_xx + iγ∂_x + ω²

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-5, 5, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=250,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Schrödinger avec potentiel cos
x, xi = symbols('x xi', real=True)
V0 = 10.0
k = 2.0

symbol = xi**2 + V0 * cos(k*x)  # -∂_xx + V₀cos(kx)

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-np.pi, np.pi, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=200,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Laplacien fractionnaire
x, xi = symbols('x xi', real=True)
alpha = 1.5  # Ordre fractionnaire

symbol = abs(xi)**alpha  # |ξ|^α

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-5, 5, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=200,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Symbole discontinu → forte non-normalité
x, xi = symbols('x xi', real=True)

# Step function in frequency
symbol = xi + 5*Heaviside(xi)  # ξ + 5H(ξ)

# Approximation lisse pour éviter les problèmes numériques
symbol_smooth = xi + 5/(1 + exp(-20*xi))

op = PseudoDifferentialOperator(symbol_smooth, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-5, 5, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=250,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Hill's operator with complex potential
x, xi = symbols('x xi', real=True)
delta = 0.5 + 0.2j  # Complex modulation

symbol = xi**2 - 2*delta*cos(2*x)  # Parametric instability

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-np.pi, np.pi, 64),
    lambda_real_range=(-5, 65),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-5, 5),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=250,
    method='spectral',
    parallel=True,
    auto_range=True
)
```

```python
# Third-order dispersion
x, xi = symbols('x xi', real=True)
beta3 = 1.0  # Dispersion coefficient

symbol = 1j * beta3 * xi**3  # iβ₃∂_xxx (Airy operator)

op = PseudoDifferentialOperator(symbol, [x], mode='symbol')
result = op.pseudospectrum_analysis(
    x_grid=np.linspace(-5, 5, 64),
    lambda_real_range=(-5, 5),  # Adjust based on expected eigenvalues
    lambda_imag_range=(-20, 20),    # Smaller range for self-adjoint operator
    epsilon_levels=[1e-1, 1e-2, 1e-3],
    resolution=250,
    method='spectral',
    parallel=True,
    auto_range=False
)
```
